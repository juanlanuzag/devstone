# Analyzing execution times of devstone models in Cadmuim Dynamic

We execute the same model using the two versions of cadmium dynamic: sequential and concurrent

```python
def result_csv_to_df(filename, val_column_name='Time'):
    measurements = pd.read_csv(filename)
    depths = set(measurements['D'])
    widths = set(measurements['W'])

    datagrid = pd.DataFrame(index=depths, columns=widths)

    for index, row in measurements.iterrows():
        datagrid[row['W']][row['D']] = row[val_column_name]
    return datagrid[datagrid.columns].astype(float)

def get_speedup_df(dfseq, dfconc):
    return dfconc.divide(dfseq)

def get_speedup_headmap(filename_seq, filename_conc):
    dfseq = result_csv_to_df(filename_seq)
    dfconc = result_csv_to_df(filename_conc)
    dfspeedup = get_speedup_df(dfseq, dfconc)
    sns.heatmap(dfspeedup, annot=True)
    plt.show()
```

Each measured time is the average of 5 runs

# Analyzing amount of atomics for devstone models


```python
def expected_atomics_LI(D, W):
    return (W-1) * (D-1) + 2 # the 2 are the eventreader and the atomic in last level
```

```python
def check_amount_of_atomics(filename):
    measurements = pd.read_csv(filename)
    for index, row in measurements.iterrows():
        if expected_atomics_LI(row['D'],row['W']) != row['AtCount']:
            print("ERROR")
            break
    print("So crack")
    
check_amount_of_atomics('modcountLI_conc.csv')
check_amount_of_atomics('modcountLI_seq.csv')
check_amount_of_atomics('modcountHI_conc.csv')
check_amount_of_atomics('modcountHI_seq.csv')
check_amount_of_atomics('modcountHO_conc.csv')
check_amount_of_atomics('modcountHO_seq.csv')
```

```python
def plot_speedup_vs_atomic_count(file_atcount, file_timeseq, file_timeconc, min_node_count=0):
    LI_atomic_count = result_csv_to_df(file_atcount, val_column_name='AtCount')
    dfseq = result_csv_to_df(file_timeseq)
    dfconc = result_csv_to_df(file_timeconc)
    dfspeedup = get_speedup_df(dfseq, dfconc)

    atomic_count = []
    speedup = []
    for row_index, row in dfspeedup.iterrows():
        for col in row.index:
            if LI_atomic_count[col][row_index] >= min_node_count:
                atomic_count.append(LI_atomic_count[col][row_index])
                speedup.append(dfspeedup[col][row_index])
    plt.plot(atomic_count, speedup, 'o', alpha=0.7)
    
def plot_speedup_vs_atomic_count_all(min_node_count=0):
    plot_speedup_vs_atomic_count('modcountLI_seq.csv', 'resultsLI_seq.csv', 'resultsLI_conc.csv', min_node_count)
    plot_speedup_vs_atomic_count('modcountHI_seq.csv', 'resultsHI_seq.csv', 'resultsHI_conc.csv', min_node_count)
    plot_speedup_vs_atomic_count('modcountHO_seq.csv', 'resultsHO_seq.csv', 'resultsHO_conc.csv', min_node_count)
    plot_speedup_vs_atomic_count('modcountHOmod_seq.csv', 'resultsHOmod_seq.csv', 'resultsHOmod_conc.csv', min_node_count)
    plt.show()
```

```python
plot_speedup_vs_atomic_count_all(0)
plot_speedup_vs_atomic_count_all(50)
```

```python
def plot_exectime_vs_atomic_count(file_atcount, file_timeseq, file_timeconc, min_node_count=0):
    LI_atomic_count = result_csv_to_df(file_atcount, val_column_name='AtCount')
    dfseq = result_csv_to_df(file_timeseq)
    dfconc = result_csv_to_df(file_timeconc)
    
    atomic_count = []
    seqtime = []
    conctime = []
    for row_index, row in dfseq.iterrows():
        for col in row.index:
            if LI_atomic_count[col][row_index] >= min_node_count:
                atomic_count.append(LI_atomic_count[col][row_index])
                seqtime.append(dfseq[col][row_index])
                conctime.append(dfconc[col][row_index])
    plt.plot(atomic_count, seqtime, 'o', alpha=0.7)
    plt.plot(atomic_count, conctime, 'o', alpha=0.7)

plot_exectime_vs_atomic_count('modcountLI_seq.csv', 'resultsLI_seq.csv', 'resultsLI_conc.csv', 0)
plt.show()
plot_exectime_vs_atomic_count('modcountHI_seq.csv', 'resultsHI_seq.csv', 'resultsHI_conc.csv', 0)
plt.show()
plot_exectime_vs_atomic_count('modcountHO_seq.csv', 'resultsHO_seq.csv', 'resultsHO_conc.csv', 0)
plt.show()
plot_exectime_vs_atomic_count('modcountHOmod_seq.csv', 'resultsHOmod_seq.csv', 'resultsHOmod_conc.csv', 0)
plt.show()
```
